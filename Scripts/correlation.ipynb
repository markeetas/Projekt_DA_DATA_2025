```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("mars_radiation_prediction.csv")
```

```python
# Select only relevant numeric columns 
columns = [
    "pressure_norm",
    "EPITHERMAL_FLUX1_norm",
    "EPITHERMAL_FLUX2_norm",
    "FAST_FLUX1_norm",
    "FAST_FLUX2_norm",
    "stormy_days_normalized",
    "elevation_m",
    "predicted_dose_e"
]
```

```python
# Compute the correlation matrix
corr = df[columns].corr()
```

```python
# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", square=True, linewidths=0.5, cbar_kws={"label": "Correlation"})
plt.title("Correlation matrix with predicted_dose_e", fontsize=14)
plt.tight_layout()
plt.show()
```
